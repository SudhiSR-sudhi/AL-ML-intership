```python
'''Generate the Fibonacci series up to N terms.'''

n = 8
series= []
a,b = 0, 1
count = 0 
while count< n:
    series.append(a)

    next_num = a+b
    a=b
    b=next_num

    count+= 1
print(series)
```

```python
'''gcd'''

a = 24
b = 18

while b != 0:
    remainder = a % b
    a = b
    b = remainder

print("GCD:", a)
```

```python
'''lcm'''

num1 = 12
num2 = 18
a = num1
b = num2

while b != 0:
    remainder = a % b
    a = b
    b = remainder

gcd = a
lcm = (num1 * num2) // gcd
print("LCM:", lcm)
```

```python
'''frequency'''

num = 5525
target = 5
count = 0

while num > 0:
    last_digit = num % 10
    if last_digit == target:
        count += 1
    num = num // 10

print("Frequency:", count)
```

```python
'''count even and odd'''

num = 45789
even = 0
odd = 0

while num > 0:
    digit = num % 10
    if digit % 2 == 0:
        even += 1
    else:
        odd += 1
    num = num // 10

print("Even:", even, "Odd:", odd)
```

```python
'''second largest digit'''

num = 2984
largest = -1
second_largest = -1

while num > 0:
    digit = num % 10
    if digit > largest:
        second_largest = largest
        largest = digit
    elif digit > second_largest and digit != largest:
        second_largest = digit
    num = num // 10

print("Second Largest:", second_largest)
```
