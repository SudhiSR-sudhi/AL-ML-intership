```python
'''Create two lists [1, 2, 3] and [4, 5, 6] and combine them into one list.'''

list1 = [1, 2, 3]
list2 = [4, 5, 6]

list1.extend(list2)

print(list1)

'''nested way'''
list1 = [1, 2, 3]
list2 = [4, 5, 6]

list_of_list = (list1,list2)
print(list_of_list)
```
