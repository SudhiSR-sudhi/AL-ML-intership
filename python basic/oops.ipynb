```python
class ATM:
    def __init__(self, initial_pin):
        self.__pin = str(initial_pin)  # Private variable

    def verify_pin(self, pin):
        """Checks if the provided pin matches the stored pin."""
        return str(pin) == self.__pin

    def change_pin(self, old_pin, new_pin):
        """Changes the pin if the old pin is correct."""
        if self.verify_pin(old_pin):
            self.__pin = str(new_pin)
            print("PIN successfully changed.")
            return True
        else:
            print("Error: Incorrect old PIN.")
            return False
        
my_atm = ATM("1234")
print(my_atm.verify_pin("1234"))
my_atm.change_pin("1234", "5678")
print(my_atm.verify_pin("1234"))
print(my_atm.verify_pin("5678"))
```

```python
class Student:

    def __init__(self, initial_marks=0):
        self.__marks = 0  # Private variable initialization
        self.set_marks(initial_marks)

    def set_marks(self, marks):
        
        if 0 <= marks <= 100:
            self.__marks = marks
        else:
            print("Error: Marks must be between 0 and 100.")

    def get_marks(self):
        return self.__marks
    
student = Student()
student.set_marks(85)
print(f"Current Marks: {student.get_marks()}")
student.set_marks(120)
print(f"Current Marks: {student.get_marks()}")

```

```python
'''Create a Login class with:

  Public variable: username
  Private variable: __password
  Method verify_password(password)
  Method change_password(old_password, new_password)'''

class UserAccount:

    def __init__(self, username, initial_password):
        self.username = username
        self.__password = str(initial_password)  # Private variable

    def verify_password(self, password):
        return str(password) == self.__password

    def change_password(self, old_password, new_password):
        if self.verify_password(old_password):
            self.__password = str(new_password)
            print("Password successfully updated.")
            return True
        else:
            print("Error: Incorrect old password.")
            return False


user = UserAccount("Alice", "abc123")
user.change_password("WrongPass", "NewSecret456")  
user.change_password("Secure123", "NewSecret456")
```

```python
class StudentPortal:

    def __init__(self, student_name, initial_cgpa):
        self.student_name = student_name  # Public variable
        self.__cgpa = float(initial_cgpa)  # Private variable

    def update_cgpa(self, cgpa):
       
        if 0.0 <= float(cgpa) <= 10.0:
            self.__cgpa = float(cgpa)
            print("CGPA successfully updated.")
            return True
        else:
            print("Error: Invalid CGPA. Must be between 0.0 and 10.0.")
            return False

    def view_profile(self):
        print(f"Name: {self.student_name}")
        print(f"CGPA: {self.__cgpa}")
        print("-----------------------")


portal = StudentPortal("Rahul Sharma", 8.5)
portal.view_profile()
portal.update_cgpa(9.1)
portal.view_profile()
portal.update_cgpa(11.5)
```
