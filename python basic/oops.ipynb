```python
class ATM:
    def __init__(self, initial_pin):
        self.__pin = str(initial_pin)  # Private variable

    def verify_pin(self, pin):
        """Checks if the provided pin matches the stored pin."""
        return str(pin) == self.__pin

    def change_pin(self, old_pin, new_pin):
        """Changes the pin if the old pin is correct."""
        if self.verify_pin(old_pin):
            self.__pin = str(new_pin)
            print("PIN successfully changed.")
            return True
        else:
            print("Error: Incorrect old PIN.")
            return False
        
my_atm = ATM("1234")
print(my_atm.verify_pin("1234"))
my_atm.change_pin("1234", "5678")
print(my_atm.verify_pin("1234"))
print(my_atm.verify_pin("5678"))
```

```python
class Student:

    def __init__(self, initial_marks=0):
        self.__marks = 0  # Private variable initialization
        self.set_marks(initial_marks)

    def set_marks(self, marks):
        
        if 0 <= marks <= 100:
            self.__marks = marks
        else:
            print("Error: Marks must be between 0 and 100.")

    def get_marks(self):
        return self.__marks
    
student = Student()
student.set_marks(85)
print(f"Current Marks: {student.get_marks()}")
student.set_marks(120)
print(f"Current Marks: {student.get_marks()}")

```
