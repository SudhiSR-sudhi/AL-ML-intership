```python
def is_palindrome(name):
   name = name.lower()
   if name == name[::-1]:
      return True
   else:
      return False
   
word = input("enter the name")
if is_palindrome(word):
   print("it is palindrome")
else:
   print("it is not")
   
```

```python
def word_frequency(name):
    name = name.lower()
    words = name.split()

    frequency = {}
    for word in words:
        if word in frequency:
            frequency[word]+=1
        else:
            frequency[word] = 1

    word_frequency("python is a high level language and python is easy to read")
    print(word_frequency)
```

```python
def remove_duplicate(list):
    unique = []
    for items in list:
        if items not in unique:
            unique.append(items) 
    return unique

numbers = [10, 20, 30, 20, 40, 10, 50]
result = remove_duplicate(numbers)

print("list", numbers)
print("reverse number", result)
```

```python
def count_letters_digits(sentence):
    letters = 0
    digits = 0

    for ch in sentence:
        if ch.isalpha():
            letters += 1
        elif ch.isdigit():
            digits += 1

    print("Letters:", letters)
    print("Digits:", digits)

text = input("enter the sentence")
count_letters_digits(text)
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact = fact*i
    return fact

number =int(input("enter the number"))
print("factorail", factorial(number))
```
