```python
'''Sort a list of tuples based on the second element (marks) using sorted() and lambda.'''

students = [("Alice", 88), ("Bob", 95), ("Charlie", 72)]
sorted_students = sorted(students, key=lambda student: student[1])
print(sorted_students)
```

```python
'''Sort a dictionary by its values using lambda. '''

scores = {"Alice": 88, "Bob": 95, "Charlie": 72}
sorted_dict = dict(sorted(scores.items(), key=lambda item: item[1]))
print(sorted_dict)
```

```python
''' Find the longest word from a list using max() and lambda.'''

words = ["apple", "banana", "watermelon", "cherry"]
longest_word = max(words, key=lambda word: len(word))
print(longest_word)
```

```python
'''Simulate a bank withdrawal and raise exceptions for insufficient balance or negative
withdrawal.'''

balance = 1000.0
try:
    amount = float(input(f"Your balance is ${balance}. Enter withdrawal amount: "))
    
    if amount <= 0:
        raise ValueError("Withdrawal amount must be a positive number.")
    if amount > balance:
        raise PermissionError("Insufficient funds for this transaction.")
        
    balance -= amount
    print(f"Withdrawal successful! Remaining balance: ${balance}")
except ValueError as e:
    print(f"Input Error: {e}")
except PermissionError as e:
    print(f"Transaction Denied: {e}")
```

```python
'''Build a calculator that handles invalid operators, invalid input, and division by zero.'''

try:
    num1 = float(input("Enter first number: "))
    operator = input("Enter operator (+, -, *, /): ").strip()
    num2 = float(input("Enter second number: "))

    if operator not in ['+', '-', '*', '/']:
        raise KeyError("Unsupported arithmetic operator.")

    if operator == '+':
        result = num1 + num2
    elif operator == '-':
        result = num1 - num2
    elif operator == '*':
        result = num1 * num2
    elif operator == '/':
        result = num1 / num2  

    print(f"Calculation Result: {result}")

except ValueError:
    print("Error: Non-numeric values entered.")
except KeyError as e:
    print(f"Error: {e}")
except ZeroDivisionError:
    print("Error: Math breakdown. Division by zero is undefined.")
```

```python
def add(x, y):
    """Performs addition."""
    return x + y

def subtract(x, y):
    """Performs subtraction."""
    return x - y

def multiply(x, y):
    """Performs multiplication."""
    return x * y

def divide(x, y):
    """Performs division and handles division by zero."""
    if y == 0:
        return "Error: Division by zero is undefined."
    return x / y

def get_numbers():
    """Helper function to safely get two numbers from the user."""
    try:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        return num1, num2
    except ValueError:
        print("Invalid input. Please enter numbers only.")
        return None

# Main program loop
while True:
    print("\n--- Simple Calculator Menu ---")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")
    
    choice = input("Enter your choice (1-5): ").strip()
    
    if choice == '1':
        nums = get_numbers()
        if nums:
            print(f"Result: {nums[0]} + {nums[1]} = {add(nums[0], nums[1])}")
            
    elif choice == '2':
        nums = get_numbers()
        if nums:
            print(f"Result: {nums[0]} - {nums[1]} = {subtract(nums[0], nums[1])}")
            
    elif choice == choice == '3':
        nums = get_numbers()
        if nums:
            print(f"Result: {nums[0]} * {nums[1]} = {multiply(nums[0], nums[1])}")
            
    elif choice == '4':
        nums = get_numbers()
        if nums:
            result = divide(nums[0], nums[1])
            if isinstance(result, str):
                print(result) # Prints the division by zero error message
            else:
                print(f"Result: {nums[0]} / {nums[1]} = {result}")
                
    elif choice == '5':
        print("Exiting calculator. Goodbye!")
        break
        
    else:
        print("Invalid choice. Please select a valid option from 1 to 5.")
```
