```python
import tkinter as tk
from tkinter import messagebox

def login():
    username = entry_username.get()
    password = entry_password.get()
    if username=="admin" and password == "password":
        messagebox.showinfo("Login","Login successfully")
    else:
        messagebox.showerror("Login" , " Invalid Login ")

root = tk.Tk()
root.title("login form")
root.geometry("300x250")
tk.Label(root,text='login form', font=('Arial',16)).pack(pady=20)
frame=tk.Frame(root)
frame.pack()

tk.Label(frame,text ="Username:").grid(row=0,column=0,pady=5)
entry_username = tk.Entry(frame)
entry_username.grid(row =0, column = 1 ,pady = 5)

tk.Label(frame,text ="password:").grid(row=1,column=0,pady=5)
entry_password = tk.Entry(frame, show ="*")
entry_password.grid(row =1, column = 1 ,pady = 5)

tk.Button(frame, text ="Login", command = login).grid(row=2, columnspan=2, pady =10)
root.mainloop()
```

```python
#calculator

import tkinter as tk


def calculate(operation):
    try:
        # Get numbers from entry fields and convert to float
        num1 = float(entry1.get())
        num2 = float(entry2.get())

        # Perform operation based on button click
        if operation == "+":
            res = num1 + num2
        elif operation == "-":
            res = num1 - num2
        elif operation == "*":
            res = num1 * num2
        elif operation == "/":
            if num2 == 0:
                result_label.config(text="Error: Division by 0", fg="red")
                return
            res = num1 / num2

        
        result_label.config(text=f"Result: {res}", fg="black")

    except ValueError:
        result_label.config(text="Error: Invalid Input", fg="red")



root = tk.Tk()
root.title("Simple Calculator")
root.geometry("300x250")


tk.Label(root, text="Number 1:").pack(pady=5)
entry1 = tk.Entry(root, width=20)
entry1.pack()

tk.Label(root, text="Number 2:").pack(pady=5)
entry2 = tk.Entry(root, width=20)
entry2.pack()

button_frame = tk.Frame(root)
button_frame.pack(pady=15)

btn_add = tk.Button(
    button_frame, text="+", width=5, command=lambda: calculate("+")
)
btn_add.grid(row=0, column=0, padx=5)

btn_sub = tk.Button(
    button_frame, text="-", width=5, command=lambda: calculate("-")
)
btn_sub.grid(row=0, column=1, padx=5)

btn_mul = tk.Button(
    button_frame, text="*", width=5, command=lambda: calculate("*")
)
btn_mul.grid(row=0, column=2, padx=5)

btn_div = tk.Button(
    button_frame, text="/", width=5, command=lambda: calculate("/")
)
btn_div.grid(row=0, column=3, padx=5)


result_label = tk.Label(root, text="Result: ", font=("Arial", 12, "bold"))
result_label.pack(pady=10)

root.mainloop()
```

```python
#login page

import tkinter as tk
from tkinter import messagebox


def check_login():
    username = user_entry.get().strip()
    password = pass_entry.get().strip()

    if username == "admin" and password == "1234":
        messagebox.showinfo("Success", "Login Successful!")
        open_home_page()
    else:
        messagebox.showerror("Error", "Invalid Username or Password")


def open_home_page():
    for widget in root.winfo_children():
        widget.destroy()

    root.title("Home Page")
    root.geometry("400x200")


    welcome_lbl = tk.Label(
        root, text="Welcome to the Home Page!", font=("Arial", 16, "bold")
    )
    welcome_lbl.pack(expand=True)


root = tk.Tk()
root.title("Login Page")
root.geometry("350x400")

tk.Label(root, text="Username:", font=("Arial", 10)).pack(pady=2)
user_entry = tk.Entry(root, width=25)
user_entry.pack(pady=2)

tk.Label(root, text="Password:", font=("Arial", 10)).pack(pady=2)
pass_entry = tk.Entry(root, width=25, show="*")
pass_entry.pack(pady=2)

login_btn = tk.Button(
    root,
    text="Login",
    command=check_login,
    width=15,
    bg="#1a73e8",
    fg="white",
    font=("Arial", 10, "bold"),
)
login_btn.pack(pady=20)

root.mainloop()
```

```python
import tkinter as tk


def calculate_results():
    try:
        marks = [
            float(e_m1.get()),
            float(e_m2.get()),
            float(e_m3.get()),
            float(e_m4.get()),
            float(e_m5.get()),
        ]

        if any(m < 0 or m > 100 for m in marks):
            lbl_res.config(text="Marks must be between 0 and 100", fg="red")
            return

        total = sum(marks)
        average = total / 5
        percentage = (total / 500) * 100

        # Conditional Grade Logic
        if percentage >= 90:
            grade = "A+"
        elif percentage >= 80:
            grade = "A"
        elif percentage >= 70:
            grade = "B"
        elif percentage >= 60:
            grade = "C"
        elif percentage >= 50:
            grade = "D"
        else:
            grade = "F"

        summary_text = (
            f"Total Marks: {total:.1f} / 500\n"
            f"Average: {average:.1f}\n"
            f"Percentage: {percentage:.1f}%\n"
            f"Grade: {grade}"
        )
        lbl_res.config(text=summary_text, fg="green")

    except ValueError:
        lbl_res.config(text="Please enter valid numbers for all fields", fg="red")


root = tk.Tk()
root.title("Marks Calculator")
root.geometry("320x420")

labels = ["Subject 1:", "Subject 2:", "Subject 3:", "Subject 4:", "Subject 5:"]
entries = []

for i, label_text in enumerate(labels):
    tk.Label(root, text=label_text).grid(
        row=i, column=0, padx=20, pady=8, sticky="w"
    )
    e = tk.Entry(root, width=15)
    e.grid(row=i, column=1, padx=20, pady=8)
    entries.append(e)

e_m1, e_m2, e_m3, e_m4, e_m5 = entries

btn_calc = tk.Button(
    root,
    text="Calculate Performance",
    command=calculate_results,
    bg="blue",
    fg="white",
)
btn_calc.grid(row=5, column=0, columnspan=2, pady=15)

lbl_res = tk.Label(root, text="", font=("Arial", 11, "bold"), justify="left")
lbl_res.grid(row=6, column=0, columnspan=2, pady=10)

root.mainloop()
```

```python
import time
import tkinter as tk


def update_time():
    clock_label.config(text=time.strftime("%H:%M:%S"))

    clock_label.after(1000, update_time)


root = tk.Tk()
root.title("Clock")

clock_label = tk.Label(root, font=("Arial", 30))
clock_label.pack(padx=20, pady=20)

update_time()

root.mainloop()
```
