```python
n = [ 'hello' , 'man' , 'hai']

for i in n: 
    print(i)
```

```python
n = [ 'hello' , 'man' , 'hai']

for i in n: 
    print(i[1:])
```

```python
n = [ 'hello' , 'man' , 'hai']

for i in n [::-1]: 
    print (f'printed {i}')
```

```python
'''example 2'''

list_numbers = [1,2,4,7,8,5,9]

for numbers in list_numbers:
    square_number = numbers**2
    print(f"square of number{numbers} is {square_number}")
```
