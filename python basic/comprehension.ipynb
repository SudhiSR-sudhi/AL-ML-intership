```python
number = [2,3,5,8,9]
square_number = [num**2 for num in number]
print(square_number)
```

```python
'''even condition'''

number = [2,3,5,8,9]
square_number = [num**2 for num in number if num%2 == 0]
print(square_number)
```

```python
'''odd condition'''

number = [2,3,5,8,9]
square_number = [num**2 for num in number if num%2 != 0]
print(square_number)
```

```python
number  = [2,3,4,5,6,7,8,9]
square_number = [num**2 if num%2 == 0 else num**3 for num in number]
print(square_number)
```

```python
list_words = ["blue", "red", "yellow", "green"]

character_count = [len(words) for words in list_words]
print(character_count)
```

```python
list_words = ["blue", "red", "yellow", "green"]

capital_letter = [words.upper() for words in list_words]
print(capital_letter)
```

```python
numbers_with = [num for num in range(0,101) if num %5 == 0 and num% 3 == 0]
print(numbers_with)
```

```python
list_words = ["blue", "red", "yellow", "green"]
first_capital = [words.capitalize() for words in list_words]
print(first_capital)
```

```python
sentence = "i love india"

vowels = [sent  for sent in sentence if sent.lower() in 'aeiou' ]
print(vowels)
```

```python
sentence = "i love india"

words = [sent for sent in sentence.split() if len(sent)>3]
print(words)
```
