```python
'''Create two sets {1, 2, 3} and {3, 4, 5} and print their union.'''

set1 = {1, 2, 3}
set2 = {3, 4, 5}

set_full = set1|set2
print(set_full)
```

```python
'''Write a program to print the common elements from the sets {1, 2, 3} and {2, 3, 4}'''

set1 = {1, 2, 3}
set2 = {2, 3, 4}

set1.intersection(set2)
```
