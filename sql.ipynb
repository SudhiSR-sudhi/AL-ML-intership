```python
import sqlite3

conn = sqlite3.connect("employee.db")
cursor = conn.cursor()
cursor.execute(''' 
    CREATE TABLE IF NOT EXISTS employee (
              id INTEGER PRIMARY KEY AUTOINCREMENT,
              name TEXT NOT NULL,
              age INTEGER NOT NULL,
              Address TEXT NOT NULL  
               )
        ''')
cursor.execute("INSERT INTO employee (name,age,Address) VALUES(?,?,?)", ("sudhi",24,"Swathy punnamkonam kachani TVMP"))

cursor.execute("INSERT INTO employee (name,age,Address) VALUES(?,?,?)", ("sunny",25,"Karthika nettayam TVMP"))

cursor.execute("INSERT INTO employee (name,age,Address) VALUES(?,?,?)", ("sooraj",28,"Arun bhavan Vattiyoorkavu TVMP "))

conn.commit()

cursor.execute("SELECT * FROM employee")
rows = cursor.fetchall()

print(" employee records ")
for row in rows:
    print(f"ID: {row[0]}, name: {row[1]}, age: {row[2]} Address:{row[3]}")

conn.close()
```
